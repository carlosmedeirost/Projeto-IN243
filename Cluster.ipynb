```
import seaborn as sns
import pandas as pd
```

```
dataset1 = pd.read_csv('https://raw.githubusercontent.com/lucheol/in243/master/projeto/cluster_dataset_1.csv')
dataset1.head()
```

```
dataset2 = pd.read_csv('https://raw.githubusercontent.com/lucheol/in243/master/projeto/cluster_dataset_2.csv')
dataset2.head()
```

```
dataset3 = pd.read_csv('https://raw.githubusercontent.com/lucheol/in243/master/projeto/cluster_dataset_3.csv')
dataset3.head()
```

```
dataset1.drop(['Unnamed: 0'], axis=1, inplace=True)
dataset2.drop(['Unnamed: 0'], axis=1, inplace=True)
dataset3.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_dataset1 = scaler.fit_transform(dataset1)
scaled_dataset2 = scaler.fit_transform(dataset2)
scaled_dataset3 = scaler.fit_transform(dataset3)
```

```
from sklearn.cluster import KMeans
def get_kmeans(numero_clusters, dataset):
    model = KMeans(n_clusters=numero_clusters)
    model.fit(dataset)
    return [numero_clusters, model.inertia_]
```

```
get_kmeans(20, scaled_dataset1)
```

```
get_kmeans(20, scaled_dataset2)
```

```
get_kmeans(20, scaled_dataset3)
```

```
resultado_dataset1 = [get_kmeans(numero_de_grupos, scaled_dataset1) for numero_de_grupos in range(1,21)]
resultado_dataset2 = [get_kmeans(numero_de_grupos, scaled_dataset2) for numero_de_grupos in range(1,21)]
resultado_dataset3 = [get_kmeans(numero_de_grupos, scaled_dataset3) for numero_de_grupos in range(1,21)]
```

```
df_resultado_dataset1 = pd.DataFrame(resultado_dataset1, columns = ['groups', 'inertia'])
df_resultado_dataset2 = pd.DataFrame(resultado_dataset2, columns = ['groups', 'inertia'])
df_resultado_dataset3 = pd.DataFrame(resultado_dataset3, columns = ['groups', 'inertia'])
```

```
plt.plot(df_resultado_dataset2['groups'], df_resultado_dataset2['inertia'])
```

```
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2)
model.fit(scaled_dataset3)
model.labels_
```

```
df_groups = pd.DataFrame(model.cluster_centers_, columns=dataset3.columns)
df_groups.head()
```

```
df_groups.transpose().plot.bar(subplots=True, figsize=(25,25), sharex=False, rot=1)
```
